# Q1 What are the steps for handling missing data?

```python

import numpy as np
dic ={"Name":["Altaf","Imran","Hamza","Bahtawar","RAnveer","Kaif","Deepika","Mahira","Waseem"],
        "State":[np.nan,"London","Lahore","Peshawar","Dehli","Mumbai","Lahore","Karachi",np.nan],
          "Gender":["M","M","M","F","M",np.nan,"F",np.nan,"M"],
          "Age":["63","33",np.nan,"19","34","32","31","38",np.nan]
     
     }
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(dic)
```

```python
df
```

```python
##Dropping rows for missing values
df.dropna()
```

```python
# droping a columns which have atleast 1 missing values
df.dropna(axis = 1)
```

```python
#droping if all data missing
df.dropna(how = 'all')
```

```python
# replacing null values in Gender with No gender with M
df["Gender"].fillna("M", inplace = True)
df
```
