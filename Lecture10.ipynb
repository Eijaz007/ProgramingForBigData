```python
cat=['low',"av","hi"]
df.replace("?", np.nan, inplace=True)
m=df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, m, inplace=True)
```

```python
df['horsepower']=df["horsepower"].astype("int")
s=np.linspace(df["horsepower"].min(),df["horsepower"].max(),4)
```

```python
print(s)
df["horsepower"].head()
```

```python
df["horsepower-bin"]=pd.cut(df["horsepower"],s,labels=cat,include_lowest=True)
```
